```python
import pandas as pd
```

```python
# Compile datasets into a single .csv file
files = ['JC-202201-citibike-tripdata.csv',
         'JC-202202-citibike-tripdata.csv',
         'JC-202203-citibike-tripdata.csv',
         'JC-202204-citibike-tripdata.csv',
         'JC-202205-citibike-tripdata.csv',
         'JC-202206-citibike-tripdata.csv',
         'JC-202207-citbike-tripdata.csv',
         'JC-202208-citibike-tripdata.csv',
         'JC-202209-citibike-tripdata.csv',
         'JC-202210-citibike-tripdata.csv',
         'JC-202211-citibike-tripdata.csv',
         'JC-202212-citibike-tripdata.csv']
        
```

```python
#Create an new DF to house the data
roop_bike_df = pd.DataFrame()
```

```python
# Loop through files to compile data
for file in files:
    data = pd.read_csv(file)
    roop_bike_df = pd.concat([roop_bike_df, data], axis=0)
```

```python
# Show new DF
roop_bike_df
```

```python
# Dropping rows with missing values
roop_bike_df.dropna(subset = ['end_station_id', 'end_lat', 'end_lng'], inplace=True)
```

```python
# Making new Start Date and Time columns
roop_bike_df[['Start Date', 'Start Time']] = roop_bike_df['started_at'].str.split(' ', expand=True)
```

```python
# Making new End Date and Time columns
roop_bike_df[['End Date', 'End Time']] = roop_bike_df['ended_at'].str.split(' ', expand=True)
```

```python
# Dropping old columns
roop_bike_df.drop(columns=['started_at', 'ended_at'], inplace=True)
```

```python
# Checking for updates to row count
roop_bike_df
```

```python
cleaned_roop_df = roop_bike_df.rename(columns={"ride_id":"ID", "rideable_type":"Bike Type", "started_at":"Start Time",
                                               "ended_at":"End Time", "start_station_name":"Start Station", "start_station_id":"Start Station ID",
                                               "end_station_name":"End Station Name", "end_station_id":"End Station ID", "start_lat":"Latitude",
                                               "start_lng":"Longitude", "end_lat":"Latitude1", "end_lng":"Longitude1", "member_casual":"Rental type" })
```

```python
# Creating output
cleaned_roop_df.to_csv("cleaned_roop_bike_data.csv", index=False, header=True)
```
